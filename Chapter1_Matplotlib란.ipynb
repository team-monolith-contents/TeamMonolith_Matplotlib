# **Chapter 1 Matplotlib란?**

### **1. Matplotlib이란?**

> Matplotlib은 1~3차원 데이터를 시각화하여 사람이 볼 수 있도록 하는 파이썬 라이브러리입니다.</br></br>
  Matplotlib을 이용하면 꺾은선, 막대 그래프, 히스토그램 등 다양한 유형의 그래프를 간편하게 생성할 수 있습니다. </br></br>
  또한 곡선의 종류, 눈금, 그래프의 제목 등 다양한 시각적 요소를 원하는 방식으로 수정할 수 있습니다.

이제부터 위 그림과 같이 각 데이터에 알맞은 다양한 그래프를 그려보도록 하겠습니다.
> 먼저, Numpy나 Pandas와 같은 다른 패키지처럼 import를 활용하여 Matplotlib을 불러옵니다.

```python
import matplotlib.pyplot as plt  # matplotlib.pyplot 를 plt라는 이름으로 축약하여 불러옵니다. 
```

> 위 import 명령어를 보면 matplotlib이 아닌, matplotlib.pyplot을 불러오고 있습니다.   
matplotlib은 파이썬에서 간단하게 그래프를 표현할 수 있는 명령어들을 pyplot라는 모듈에 넣어놓았습니다.   
따라서 그래프를 그릴 때에는 matplotlib 중에서도 pyplot만을 특정하여 불러올 수 있습니다. </br></br>
  이 모듈은 일반적으로 plt라고 줄여서 사용하고 있습니다.

### **2. Matplotlib.pyplot 사용법**
1차원 list 혹은 array를 1차원 선그래프로 표현하는 함수에는 대표적으로 `plt.plot()`, `plt.show()` 가 있습니다. </br>

> 활용 방법 </br>
  - plt.plot(원하는 데이터) : 1차원 선 그래프를 생성합니다. </br>
  - plt.show() : 생성된 그래프를 모니터 화면에 출력합니다.

```python
# 예제 1 : plt.plot을 활용하여 1차원 그래프 그리기

data = [1,2,3,5,4]  # list 데이터를 저장합니다.
plt.plot(data) # 1차원 선그래프를 생성합니다. 
plt.show() # 앞서 생성한 그래프를 모니터 화면에 출력합니다. 
```

> pyplot의 함수에 숫자 배열 'data'를 입력하여 위와 같은 그래프가 생성되었습니다.   
만들어진 그래프를 보면, plt.plot() 함수는 입력된 값들을 y값으로 가정하고, x값([0, 1, 2, 3])을 자동으로 만들어낸다는 것을 알 수 있습니다. 

### **3. x축, y축 값을 직접 입력하여 그래프를 생성하기**

> x값을 직접 입력하는 방법도 있습니다.  
> 아래 그림처럼 x축의 값을 직접 넣어서 그래프를 생성하면 x축의 값이 설정됩니다. 
> 활용 방법 </br>
  - plt.plot(원하는 x축 값, 원하는 y축 값)

```python
# 예제 2 : plt.plot함수에 x, y값을 직접 입력으로 주어 1차원 그래프를 그리기 

x = [-2,-1,0,1,2]
y = [1,2,3,5,4]
plt.plot(x, y)
plt.show()
```

### **4. 그래프의 속성값을 설정하여 생성하기**

- [하단 좌측 그림] pyplot.plot로 다양한 옵션을 설정하여 그래프를 생성할 수도 있습니다.
> 활용 방법1 : 포맷 문자열</br>
x, y값 입력에 대한 선의 색상과 형태를 지정하는 포맷 문자열(Format string)을 옵션으로 입력할 수 있습니다.
포맷 문자열 ‘ro’는 빨간색(‘red’)의 원형(‘o’) 마커를 의미합니다.
또한, 예를 들어 ‘b-‘는 파란색(‘blue’)의 점선(‘--‘)을 의미합니다.</br> </br> 
> 활용 방법2 : x,y축의 범위 지정</br> 
pyplot 모듈의 axis() 함수를 이용해서 축의 범위([xmin, xmax, ymin, ymax])를 지정할 수 있습니다.

- [하단 우측 그림] pyplot.plot은 여러 데이터를 한번에 입력 받아 그래프를 생성할 수 있습니다.
> 활용 방법 : 시각화하고자하는 데이터를 plot() 함수에 차례대로 입력합니다. 


```python
# 예제 3 : 포맷 문자열을 이용하여 다양한 형태의 그래프 생성

# x, y 데이터 생성
x = [-2,-1,0,1,2]
y = [1,2,3,5,4]

# 빨간색 점
plt.plot(x, y, 'ro')  # format string : 'r'(빨간색), 'o'(원형 마커)
plt.show()

# 파란색 점선
plt.plot(x, y, 'b--')  # format string : 'b'(파란색), '--'(점선)
plt.show()

# 초록색 삼각형
plt.plot(x, y, 'g^')  # format string : 'g'(초록색), '^'(삼각형 마커)
plt.show()
```

```python
##### 예제 4 : x,y축의 범위 지정

# x, y 데이터 생성
x = [-2,-1,0,1,2]
y = [1,2,3,5,4]

# 범위 지정
x_min = -3
x_max = 3
y_min = 0
y_max = 6

# 빨간색 점
plt.plot(x, y, 'ro')
plt.axis([x_min, x_max, y_min, y_max])  # 축의 범위를 지정합니다. 
plt.show()

# 파란색 점선
plt.plot(x, y, 'b--')
plt.axis([x_min, x_max, y_min, y_max])
plt.show()
```

```python
# 예제 5 : 여러 데이터를 한번에 표현하기
import numpy as np

# x, y 데이터 생성
x = np.array([-2,-1,0,1,2])
y = np.array([1,2,3,5,4])
x_2 = np.array([-3,-1,0,3,4])
y_2 = np.array([-1,0,-1,2,1])

plt.plot(x, y, 'r--', x_2, y_2, 'g^') #x, y 데이터는 빨간 점선, x_2, y_2 데이터는 초록 삼각형 마커로 그래프를 각각 그려줍니다.
plt.show()
```

### **5. 막대그래프 그리기**

> 데이터에 따라 1차원 실선 그래프가 아닌 막대 그래프가 적절할 수 있습니다. 다음 예제에서는 막대 그래프를 활용하여 그래프를 생성해보겠습니다.

> 활용 방법 : pyplot 모듈의 bar() 함수를 활용합니다. plt.bar(x축 레이블, 데이터)로 그래프를 생성할 수 있습니다.

```python
# 예제 6-1 : 막대그래프 그리기

x = ['apple', 'banana', 'kiwi'] # x축 데이터를 저장합니다. 
data = [1100, 1400, 1200] # y축 데이터를 저장합니다. 

plt.bar(x , data)
plt.show()
```

> 선 그래프와 마찬가지로 색상을 바꿀 수 있습니다. 각 막대마다 따로따로 색상을 지정할 수도 있습니다. </br>
> 활용 예시) plt.bar(x, data, color=['막대1 색상', '막대2 색상', '막대3 색상', ...])  </br>

<참고>일부 막대의 색을 지정하지 않은 경우 처음 입력한 색으로 통일됩니다.   
예를 들어 세 개의 막대 색상을 ['r']로만 입력한 경우 모든 막대의 색이 빨간색으로 통일됩니다.   
또 ['r', 'g']라고 지정한 경우, 따로 초록색으로 지정한 두번째 막대만 초록색으로 출력되고, 나머지 막대는 모두 처음 입력한 빨간색으로 통일됩니다. 

```python
# 예제 6-2 : 막대그래프 그리기

x = ['apple', 'banana', 'kiwi']
data = [1100, 1400, 1200]

plt.bar(x , data, color=['r', 'g', 'b'])
plt.show()
```

### **6. 산점도(Scatter Plot) 그래프 그리기**

> 산점도는 두 변수의 상관 관계를 점으로 표현하는 그래프입니다.</br>

> 활용 방법 : pyplot 모듈의 scatter() 함수를 활용합니다. plt.scatter(x 값, y 값)로 그래프를 생성할 수 있습니다.

```python
# 예제 7 : 산점도 그래프 그리기
import numpy as np # 배열을 활용하기 위해 numpy를 np로 축약하여 불러옵니다. 

x = np.arange(0, 4, 0.1) # numpy의 arange함수를 활용하여 0부터 4까지 0.1의 간격으로 이루어진 배열을 생성합니다. 
y = 3*x + np.random.randn(len(x)) # np.random.randn 함수는 입력으로 받은 값의 개수만큼 난수를 생성합니다. 생성된 난수에 3*x을 더하여 y값을 생성합니다.

plt.scatter(x , y)
plt.show()
```

> scatter() 함수의 s, c 파라미터를 통해 각각 마커의 크기(size)와 색상(color)을 지정할 수 있습니다. 마커의 크기는 `size**2` 의 형태로 지정합니다. </br>
예를 들어 plot() 함수에 `markersize=20`으로 지정힌 것과 scatter() 함수에 `s=20**2`으로 지정한 마커는 동일한 크기입니다.</br></br>
> 마커의 색상은 데이터의 길이와 같은 크기의 숫자 시퀀스 또는 rgb, 그리고 Hex code 색상을 입력해서 지정할 수 있습니다.   
참고로, plot함수의 markersize 란 plot 함수에 의해 그래프에 데이터를 표시하는 점의 크기를 뜻합니다. 색상 및 모양 지정에 대해서는 다음 챕터에서 상세히 다루겠습니다. 

```python
# 예제 8-1 : 산점도 그래프를 s, c 인자를 받아 그리기

# Numpy의 random.randn을 활용하여 난수를 n개 생성
n = 50
x = np.random.rand(n) # n, 즉 50개의 난수를 생성합니다. 
y = np.random.rand(n)

area = (30 * np.random.rand(n))**2  # n개의 데이터 각각의 마커 사이즈를 난수를 생성하고, 생성된 난수에 30을 곱한 값만큼의 크기로 지정할 겁니다. 
colors = np.random.rand(n) # 색상도 n개 데이터 마다 난수를 생성하여 지정합니다. 

plt.scatter(x, y, s=area, c=colors) # 위에서 변수에 저장한 값들을 이용하여 x, y, size, color가 랜덤으로 입력되는 산점도를 그립니다. 
plt.show()
```

```python
# 예제 8-2 : 산점도 그래프와 plot함수의 s인자와 markersize 인자의 크기 비교하기

plt.plot([1], [1], 'o', markersize=20)
plt.scatter([2], [1], s=20**2)
plt.axis([-1, 4, 0.8, 1.2])
```

> 위에서 언급한 것과 같이 plot() 함수의 markersize=20와 scatter() 함수의 s=20**2이 동일한 크기임을 확인할 수 있습니다. 

- 산점도 그래프는 크기뿐만 아니라 색상도 매우 중요한 요소입니다. 아래 예제에서는 alpha, cmap이라는 인자를 통해 이를 조절해보겠습니다.
> alpha 파라미터는 마커의 투명도를 지정합니다. 0에서 1 사이의 값을 입력합니다. </br>
> cmap 파라미터는 Matplotlib에서 지원하는 컬러맵에 해당하는 문자열을 입력으로 지정할 수 있습니다.

```python
# 예제 8-3 : Matplotlib의 colormap을 활용하여 산점도 그래프 꾸미기

plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Spectral')
plt.colorbar()
plt.show()
```

> cmap에 적용할 수 있는 colormap 코드의 예시입니다. 이를 활용하여 위의 예제와 다른 colormap을 한번 더 그려보겠습니다.

```python
# 예제 8-4 : Matplotlib의 colormap 코드 활용

plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='tab20c')
plt.colorbar()
plt.show()
```
