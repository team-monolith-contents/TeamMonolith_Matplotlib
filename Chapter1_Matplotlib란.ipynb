- [í•˜ë‹¨ ì¢Œì¸¡ ê·¸ë¦¼] pyplot.plotë¡œ ë‹¤ì–‘í•œ ì˜µì…˜ì„ ì„¤ì •í•˜ì—¬ ê·¸ëž˜í”„ë¥¼ ìƒì„±í•  ìˆ˜ë„ ìžˆìŠµë‹ˆë‹¤.
> í™œìš© ë°©ë²•1 : í¬ë§· ë¬¸ìžì—´</br>
x, yê°’ ìž…ë ¥ì— ëŒ€í•œ ì„ ì˜ ìƒ‰ìƒê³¼ í˜•íƒœë¥¼ ì§€ì •í•˜ëŠ” í¬ë§· ë¬¸ìžì—´(Format string)ì„ ì˜µì…˜ìœ¼ë¡œ ìž…ë ¥í•  ìˆ˜ ìžˆìŠµë‹ˆë‹¤.
í¬ë§· ë¬¸ìžì—´ â€˜roâ€™ëŠ” ë¹¨ê°„ìƒ‰(â€˜redâ€™)ì˜ ì›í˜•(â€˜oâ€™) ë§ˆì»¤ë¥¼ ì˜ë¯¸í•©ë‹ˆë‹¤.
ë˜í•œ, ì˜ˆë¥¼ ë“¤ì–´ â€˜b-â€˜ëŠ” íŒŒëž€ìƒ‰(â€˜blueâ€™)ì˜ ì ì„ (â€˜--â€˜)ì„ ì˜ë¯¸í•©ë‹ˆë‹¤.</br> </br> 
> í™œìš© ë°©ë²•2 : x,yì¶•ì˜ ë²”ìœ„ ì§€ì •</br> 
pyplot ëª¨ë“ˆì˜ axis() í•¨ìˆ˜ë¥¼ ì´ìš©í•´ì„œ ì¶•ì˜ ë²”ìœ„([xmin, xmax, ymin, ymax])ë¥¼ ì§€ì •í•  ìˆ˜ ìžˆìŠµë‹ˆë‹¤.

- [í•˜ë‹¨ ìš°ì¸¡ ê·¸ë¦¼] pyplot.plotì€ ì—¬ëŸ¬ ë°ì´í„°ë¥¼ í•œë²ˆì— ìž…ë ¥ ë°›ì•„ ê·¸ëž˜í”„ë¥¼ ìƒì„±í•  ìˆ˜ ìžˆìŠµë‹ˆë‹¤.
> í™œìš© ë°©ë²• : ì‹œê°í™”í•˜ê³ ìží•˜ëŠ” ë°ì´í„°ë¥¼ plot() í•¨ìˆ˜ì— ì°¨ë¡€ëŒ€ë¡œ ìž…ë ¥í•©ë‹ˆë‹¤. 


```python
# ì˜ˆì œ 3 : í¬ë§· ë¬¸ìžì—´ì„ ì´ìš©í•˜ì—¬ ë‹¤ì–‘í•œ í˜•íƒœì˜ ê·¸ëž˜í”„ ìƒì„±

# x, y ë°ì´í„° ìƒì„±
x = [-2,-1,0,1,2]
y = [1,2,3,5,4]

# ë¹¨ê°„ìƒ‰ ì 
plt.plot(x, y, 'ro')  # format string : 'r'(ë¹¨ê°„ìƒ‰), 'o'(ì›í˜• ë§ˆì»¤)
plt.show()

# íŒŒëž€ìƒ‰ ì ì„ 
plt.plot(x, y, 'b--')  # format string : 'b'(íŒŒëž€ìƒ‰), '--'(ì ì„ )
plt.show()

# ì´ˆë¡ìƒ‰ ì‚¼ê°í˜•
plt.plot(x, y, 'g^')  # format string : 'g'(ì´ˆë¡ìƒ‰), '^'(ì‚¼ê°í˜• ë§ˆì»¤)
plt.show()
```

```python
##### ì˜ˆì œ 4 : x,yì¶•ì˜ ë²”ìœ„ ì§€ì •

# x, y ë°ì´í„° ìƒì„±
x = [-2,-1,0,1,2]
y = [1,2,3,5,4]

# ë²”ìœ„ ì§€ì •
x_min = -3
x_max = 3
y_min = 0
y_max = 6

# ë¹¨ê°„ìƒ‰ ì 
plt.plot(x, y, 'ro')
plt.axis([x_min, x_max, y_min, y_max])  # ì¶•ì˜ ë²”ìœ„ë¥¼ ì§€ì •í•©ë‹ˆë‹¤. 
plt.show()

# íŒŒëž€ìƒ‰ ì ì„ 
plt.plot(x, y, 'b--')
plt.axis([x_min, x_max, y_min, y_max])
plt.show()
```

### **6. ì‚°ì ë„(Scatter Plot) ê·¸ëž˜í”„ ê·¸ë¦¬ê¸°**

> ì‚°ì ë„ëŠ” ë‘ ë³€ìˆ˜ì˜ ìƒê´€ ê´€ê³„ë¥¼ ì ìœ¼ë¡œ í‘œí˜„í•˜ëŠ” ê·¸ëž˜í”„ìž…ë‹ˆë‹¤.</br>

> í™œìš© ë°©ë²• : pyplot ëª¨ë“ˆì˜ scatter() í•¨ìˆ˜ë¥¼ í™œìš©í•©ë‹ˆë‹¤. plt.scatter(x ê°’, y ê°’)ë¡œ ê·¸ëž˜í”„ë¥¼ ìƒì„±í•  ìˆ˜ ìžˆìŠµë‹ˆë‹¤.

```python
# ì˜ˆì œ 7 : ì‚°ì ë„ ê·¸ëž˜í”„ ê·¸ë¦¬ê¸°
import numpy as np # ë°°ì—´ì„ í™œìš©í•˜ê¸° ìœ„í•´ numpyë¥¼ npë¡œ ì¶•ì•½í•˜ì—¬ ë¶ˆëŸ¬ì˜µë‹ˆë‹¤. 

x = np.arange(0, 4, 0.1) # numpyì˜ arangeí•¨ìˆ˜ë¥¼ í™œìš©í•˜ì—¬ 0ë¶€í„° 4ê¹Œì§€ 0.1ì˜ ê°„ê²©ìœ¼ë¡œ ì´ë£¨ì–´ì§„ ë°°ì—´ì„ ìƒì„±í•©ë‹ˆë‹¤. 
y = 3*x + np.random.randn(len(x)) # np.random.randn í•¨ìˆ˜ëŠ” ìž…ë ¥ìœ¼ë¡œ ë°›ì€ ê°’ì˜ ê°œìˆ˜ë§Œí¼ ë‚œìˆ˜ë¥¼ ìƒì„±í•©ë‹ˆë‹¤. ìƒì„±ëœ ë‚œìˆ˜ì— 3*xì„ ë”í•˜ì—¬ yê°’ì„ ìƒì„±í•©ë‹ˆë‹¤.

plt.scatter(x , y)
plt.show()
```

> scatter() í•¨ìˆ˜ì˜ s, c íŒŒë¼ë¯¸í„°ë¥¼ í†µí•´ ê°ê° ë§ˆì»¤ì˜ í¬ê¸°(size)ì™€ ìƒ‰ìƒ(color)ì„ ì§€ì •í•  ìˆ˜ ìžˆìŠµë‹ˆë‹¤. ë§ˆì»¤ì˜ í¬ê¸°ëŠ” `size**2` ì˜ í˜•íƒœë¡œ ì§€ì •í•©ë‹ˆë‹¤. </br>
ì˜ˆë¥¼ ë“¤ì–´ plot() í•¨ìˆ˜ì— `markersize=20`ìœ¼ë¡œ ì§€ì •ížŒ ê²ƒê³¼ scatter() í•¨ìˆ˜ì— `s=20**2`ìœ¼ë¡œ ì§€ì •í•œ ë§ˆì»¤ëŠ” ë™ì¼í•œ í¬ê¸°ìž…ë‹ˆë‹¤.</br></br>
> ë§ˆì»¤ì˜ ìƒ‰ìƒì€ ë°ì´í„°ì˜ ê¸¸ì´ì™€ ê°™ì€ í¬ê¸°ì˜ ìˆ«ìž ì‹œí€€ìŠ¤ ë˜ëŠ” rgb, ê·¸ë¦¬ê³  Hex code ìƒ‰ìƒì„ ìž…ë ¥í•´ì„œ ì§€ì •í•  ìˆ˜ ìžˆìŠµë‹ˆë‹¤.   
ì°¸ê³ ë¡œ, plotí•¨ìˆ˜ì˜ markersize ëž€ plot í•¨ìˆ˜ì— ì˜í•´ ê·¸ëž˜í”„ì— ë°ì´í„°ë¥¼ í‘œì‹œí•˜ëŠ” ì ì˜ í¬ê¸°ë¥¼ ëœ»í•©ë‹ˆë‹¤. ìƒ‰ìƒ ë° ëª¨ì–‘ ì§€ì •ì— ëŒ€í•´ì„œëŠ” ë‹¤ìŒ ì±•í„°ì—ì„œ ìƒì„¸ížˆ ë‹¤ë£¨ê² ìŠµë‹ˆë‹¤. 

```python
# ì˜ˆì œ 8-1 : ì‚°ì ë„ ê·¸ëž˜í”„ë¥¼ s, c ì¸ìžë¥¼ ë°›ì•„ ê·¸ë¦¬ê¸°

# Numpyì˜ random.randnì„ í™œìš©í•˜ì—¬ ë‚œìˆ˜ë¥¼ nê°œ ìƒì„±
n = 50
x = np.random.rand(n) # n, ì¦‰ 50ê°œì˜ ë‚œìˆ˜ë¥¼ ìƒì„±í•©ë‹ˆë‹¤. 
y = np.random.rand(n)

area = (30 * np.random.rand(n))**2  # nê°œì˜ ë°ì´í„° ê°ê°ì˜ ë§ˆì»¤ ì‚¬ì´ì¦ˆë¥¼ ë‚œìˆ˜ë¥¼ ìƒì„±í•˜ê³ , ìƒì„±ëœ ë‚œìˆ˜ì— 30ì„ ê³±í•œ ê°’ë§Œí¼ì˜ í¬ê¸°ë¡œ ì§€ì •í•  ê²ë‹ˆë‹¤. 
colors = np.random.rand(n) # ìƒ‰ìƒë„ nê°œ ë°ì´í„° ë§ˆë‹¤ ë‚œìˆ˜ë¥¼ ìƒì„±í•˜ì—¬ ì§€ì •í•©ë‹ˆë‹¤. 

plt.scatter(x, y, s=area, c=colors) # ìœ„ì—ì„œ ë³€ìˆ˜ì— ì €ìž¥í•œ ê°’ë“¤ì„ ì´ìš©í•˜ì—¬ x, y, size, colorê°€ ëžœë¤ìœ¼ë¡œ ìž…ë ¥ë˜ëŠ” ì‚°ì ë„ë¥¼ ê·¸ë¦½ë‹ˆë‹¤. 
plt.show()
```

```python
# ì˜ˆì œ 8-2 : ì‚°ì ë„ ê·¸ëž˜í”„ì™€ plotí•¨ìˆ˜ì˜ sì¸ìžì™€ markersize ì¸ìžì˜ í¬ê¸° ë¹„êµí•˜ê¸°

plt.plot([1], [1], 'o', markersize=20)
plt.scatter([2], [1], s=20**2)
plt.axis([-1, 4, 0.8, 1.2])
```

> ìœ„ì—ì„œ ì–¸ê¸‰í•œ ê²ƒê³¼ ê°™ì´ plot() í•¨ìˆ˜ì˜ markersize=20ì™€ scatter() í•¨ìˆ˜ì˜ s=20**2ì´ ë™ì¼í•œ í¬ê¸°ìž„ì„ í™•ì¸í•  ìˆ˜ ìžˆìŠµë‹ˆë‹¤.
